```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#linear data generating
linear_x=np.linspace(0,10,50)
linear_y=2*linear_x+5+np.random.randn(50)*2 #adding noise
```

```python
#nonlinear data generating
nonlinear_x=np.linspace(0,10,50)
nonlinear_y=2*np.sin(nonlinear_x)+np.random.randn(50)#adding noise
```

```python
#linear
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(linear_x,linear_y, color='blue', label='Data with linear relationship')
plt.plot(linear_x, 2*linear_x+5, color='red', label='True linear relationship')
plt.xlabel('X')
plt.xlabel('Y')
plt.title('Linear relationship')
plt.legend()
plt.show()


#monlinear
plt.figure(figsize=(10,5))
plt.subplot(1,2,2)
plt.scatter(nonlinear_x, nonlinear_y, color='green', label='Data with non linear relationship')
plt.plot(nonlinear_x, 2*np.sin(nonlinear_x), color='orange',label='True non linear relationship')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-linear relationship')
plt.legend()
plt.show
```

# correalition coefficient


```python
linear_coef=np.corrcoef(linear_x, linear_y)[0, 1]
nonlinear_coef=np.corrcoef(nonlinear_x,nonlinear_y)[0, 1]
```

```python
print("Correlation coefficent for linear data:", linear_coef)
print("Correlation coefficient for non linear data:", nonlinear_coef)
```

# Linear regression for linear data

```python
from sklearn.linear_model import LinearRegression
```

```python
reg1=LinearRegression()
```

```python
reg1.fit(linear_x.reshape(-1, 1), linear_y)
```

```python
predict_linear=reg1.predict(linear_x.reshape(-1,1))
```

```python
plt.plot(linear_x, predict_linear)
reg1.score(linear_x.reshape(-1,1),linear_y)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(linear_x,linear_y, color='blue', label='Data with linear relationship')
plt.plot(linear_x, predict_linear, color='red', label='Best fit line')
plt.xlabel('X')
plt.xlabel('Y')
plt.title('Linear relationship')
plt.legend()
plt.show()
```

# linear regression for non linear data

```python
reg2=LinearRegression()
```

```python
reg2.fit(nonlinear_x.reshape(-1,1), nonlinear_y)
```

```python
predict_nonlinear=reg2.predict(nonlinear_x.reshape(-1,1))
```

```python
reg2.score(nonlinear_x.reshape(-1,1),nonlinear_y)
```

```python
plt.plot(nonlinear_x, predict_nonlinear)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(nonlinear_x, nonlinear_y, color='green', label='Data with non linear relationship')
plt.plot(nonlinear_x, 2*np.sin(nonlinear_x), color='orange',label='True non linear relationship')
plt.plot(nonlinear_x, predict_nonlinear, color='black', label='Best fit line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Non-linear relationship')
plt.legend()
plt.show
```
