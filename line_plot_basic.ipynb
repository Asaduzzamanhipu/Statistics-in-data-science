```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.linspace(0, 4*np.pi,100)
```

```python
y1=np.sin(x)
y2=np.cos(x)
```

```python
plt.figure(figsize=(10,4))

plt.plot(x,y1, label='sinX', color='red', linestyle='-', linewidth='2', marker='o', markersize='7')
plt.plot(x,y2, label='cosX', color='green', linestyle='--', linewidth='2', marker='s', markersize='7')
plt.legend(loc='upper left', shadow=True)
plt.grid(True,linestyle='--',alpha=0.9)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.savefig('line_plot.png')
plt.show()
```
