```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew
import seaborn as sns
```

Right skewed distribution

```python
#data generation
data= np.random.lognormal(mean=0, sigma=1, size=1000)
```

```python
mean= np.mean(data)
median=np.median(data)
mode=max(set(data),key=list(data).count)
```

```python
plt.figure(figsize=(10,4))
plt.hist(data, bins=40, color='blue',label='Data')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean:{mean:.2f}')
plt.axvline(median, color='green',linestyle='dashed', linewidth=2, label=f'Median:{median:.2f}')
plt.axvline(mode,color='yellow', linestyle='dashed', linewidth=2, label=f'Mode:{mode:.2f}')


plt.title('Right skewed distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
skewness=skew(data)
print(f'Skewness:{skewness}')
```

Normal distribution where skewness is approximately zero

```python
#data generation
data= np.random.normal(loc=0, scale=1, size=1000)
```

```python
mean= np.mean(data)
median=np.median(data)
mode=max(set(data),key=list(data).count)
```

```python
plt.figure(figsize=(10,4))
plt.hist(data, bins=40, color='grey',label='Data')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean:{mean:.2f}')
plt.axvline(median, color='green',linestyle='dashed', linewidth=2, label=f'Median:{median:.2f}')
plt.axvline(mode,color='yellow', linestyle='dashed', linewidth=2, label=f'Mode:{mode:.2f}')


plt.title('Normal skewed distribution where skewdness~0')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
skewness=skew(data)
print(f'Skewness:{skewness}')
```
