boxplot using matplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.array([1,2,3,3,5,7,7,8,9,10,20])
```

```python
q1=np.percentile(data,25) #calculatile 25% or 1st percentile
q3=np.percentile(data,75)
iqr=q3-q1

lower_whisker=q1-(1.5*iqr)
upper_whisker=q3+(1.5*iqr)

outliers=(data<lower_whisker)|(data>upper_whisker)

```

```python
#creating boxplo

fig,ax=plt.subplots()
ax.boxplot(data)
ax.set_title('Boxplot example')
ax.set_xlabel('Data')
ax.set_ylabel('Value')
plt.show()
```

boxplot using seaborn

```python
import seaborn as sns
```

```python
data_points=[1,2,3,3,5,7,7,8,9,10,20]
```

```python
#creating boxplot
sns.boxplot(data=data_points,width=.1)
plt.title('Boxplot example')
plt.xlabel('Data')
plt.ylabel('Values')

plt.show()
```
